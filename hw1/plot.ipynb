## Plotting Curves

```python
import json
# load from json to retrieve two lists
with open('./qa-ckpt/loss_em.json', 'r') as f:
    data = json.load(f)

loss_list = data['loss']
em_list = data['em']
```

```python
loss_list
```

```python
em_list
```

```python
# plot line chart for loss and em respectively. Every data point is an epoch.
import matplotlib.pyplot as plt

import os

if not os.path.exists('./plots'):
    os.makedirs('./plots')
x = [1, 2, 3, 4, 5]

plt.plot(x, loss_list, marker='o')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Learning Curve of the loss value (per epoch)')
plt.savefig('./plots/loss.png')
plt.show()
```

```python
plt.plot(x, em_list, marker='o')
plt.xlabel('Epoch')
plt.ylabel('EM Score')
plt.title('Learning Curve of the Exact Match metric value (per epoch)')
plt.savefig('./plots/em.png')
plt.show()
```
