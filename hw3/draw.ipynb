```python
import matplotlib.pyplot as plt

# x: checkpoint step, y: mean perplexity
points = [
    (500, 3.404711522102356),
    (1000, 3.3634377546310423),
    (1500, 3.879033981800079),
    (2000, 4.872921903610229),
    (2500, 4.734406974315643)
]

x = [p[0] for p in points]
y = [p[1] for p in points]

plt.plot(x, y)
plt.xlabel('Checkpoint Step')
plt.ylabel('Mean Perplexity')
plt.title('Lora Training: Mean Perplexity vs. Checkpoint Step')
plt.savefig('lora_training.png')
```
